#Задачі прогнозування

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
import torch.nn.functional as F


# 2. Завантаження токенізатора та моделі GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()

```

```python
# 1. Вхідний текст (контекст)
prompt = "I love reading books about"
```

```python
# 3. Токенізація: перетворюємо текст у tensory, зрозумілі моделі
# return_tensors="pt" → отримуємо PyTorch тензори
inputs = tokenizer(prompt, return_tensors="pt")

# Витягуємо input_ids (ідентифікатори токенів), які подаватимемо в модель
input_ids = inputs["input_ids"]
# Наприклад: "I love reading books about" → [40, 389, 1084, 3615,  about_token_id]

# 4. Прогін через модель GPT-2 (інференс)

with torch.no_grad(): # вимикаємо обрахунок градієнтів
    outputs = model(**inputs)  # Передаємо input_ids в модель
    logits = outputs.logits    # logits — "сирі" передбачення для кожного токена

# 5. Беремо логіти тільки останнього токена
# GPT генерує по одному токену: потрібно взяти лише прогноз для останнього слова (що йде після "about")
# [0] — batch, [-1] — останній токен, [:] — всі токени з словника
next_token_logits = logits[0, -1, :]

# 6. Обчислюємо ймовірності: softmax перетворює логіти у ймовірнісну форму
# Отримаємо розподіл імовірностей по всьому словнику GPT-2 (близько 50,000 токенів)
probs = F.softmax(next_token_logits, dim=-1)

# 7. Витягуємо top-k (топ-10) найймовірніших наступних токенів
top_k = 10  # кількість найімовірніших варіантів, які ми хочемо побачити
top_k_probs, top_k_ids = torch.topk(probs, k=top_k)  # повертає ймовірності та відповідні токен-ID

# Вивід результату
print(f"\nPrompt: {prompt}")
print("\nTop predicted next tokens:")
for i in range(top_k):
    # Декодуємо токен-ID назад у текст (наприклад, 123 → 'history')
    token = tokenizer.decode([top_k_ids[i]])
    # Отримуємо ймовірність у зручному форматі (float)
    prob = top_k_probs[i].item()
    # Друкуємо: порядковий номер, токен, його ймовірність
    print(f"{i+1:2}. '{token}': {prob:.4f}")
```
